```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def function_z(x, y):
    return (x**2 + y**2)
```

```python
N = 20 

x_values = np.linspace(-100, 100, N)
y_values = np.linspace(-100, 100, N)
```

```python
X, Y = np.meshgrid(x_values, y_values)
```

```python
Z = function_z(X, Y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,cmap='turbo');
```
